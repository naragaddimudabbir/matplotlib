ðŸ¥§ Pie Charts in EDA using Matplotlib ðŸ”¹ What is a Pie Chart? A pie chart is a circular statistical plot used to show the proportions or percentages of a whole. Each slice of the pie represents a category's contribution to the total.

ðŸ”¹ When to Use Pie Charts in EDA? Pie charts are useful for:

Showing the relative proportion of categories

Quickly understanding dominant categories

Comparing a small number of categories (ideally <6)

Visualizing class balance in datasets

âœ… Best for high-level summaries âŒ Not ideal for detailed comparisons or large category counts

ðŸ”¹ Matplotlib for Pie Charts Matplotlib provides a simple method plt.pie() to draw pie charts.

ðŸ”¸ Syntax: python Copy Edit plt.pie(values, labels=categories, autopct='%1.1f%%')

```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")  # Choose any style you like
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 950]
plt.title("Career runs of Indian batsman's")
plt.pie(runs,labels=labels)
plt.show()
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels=labels, colors=colors)
plt.show()
```

```python
plt.pie(runs,labels=labels,colors=colors,
       wedgeprops={'edgecolor':'black','linewidth':2,'linestyle':'--','fill':True})
plt.show()
```

```python
explode = [0.1,0,0,0]
plt.pie(runs,labels=labels,explode=explode,colors=colors,shadow=True)
plt.title("Exploded Pie Example")
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True,
    startangle=120
)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True,
    autopct="%1.1f%%",
    startangle=100
)
plt.show()
```
