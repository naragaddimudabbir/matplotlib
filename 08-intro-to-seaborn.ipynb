```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="darkgrid")
```

```python
x=np.array([0, 2, 3, 4, 5, 6, 7, 8, 9 ,10, 60])
y=np.sin(x)
```

```python
sns.lineplot(x=x,y=y)
plt.title("number vs sin(number)")
plt.show()
```
