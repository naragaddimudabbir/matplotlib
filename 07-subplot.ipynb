```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8,9]
y1 = [i*2 for i in x]
y2 = [i**2 for i in x]
y3 = [i**0.5 for i in x]

y4 = [10 - i for i in x]

plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title("double")

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title("square")
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title("double")

plt.subplot(2,1,2)
plt.plot(x,y2)
plt.title("square")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("double")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("square")

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("square root")

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("10-x")

plt.tight_layout()
```

```python
fig,axs = plt.subplots(1 , 2, figsize=(12,7))
axs[0].plot(x,y1)
axs[0].set_title("double")
axs[1].plot(x,y2)
axs[1].set_title("square")
fig.tight_layout()


plt.show()
                   
                      
```

```python
fig,axs=plt.subplots(2,2,figsize=(12,7))

ys=[y1,y2,y3,y4]
title = ["double","square","square root","10-x"]
for i in range(2):
    for j in range(2):
        idx = i*2+j
        axs[i,j].plot(x,ys[idx])
        axs[i,j].set_title(title[idx])
plt.tight_layout()
plt.show()
        
```
